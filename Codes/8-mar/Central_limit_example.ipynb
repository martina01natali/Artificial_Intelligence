# Central limit theorem

- produce N independent random variables distributed with a flat distribution
- compute the sample's mean for k<N variables
- show the "gaussianization" of the mean when k grows

```python
import numpy as np
import random
import matplotlib
from matplotlib import pyplot as plt
```

```python
sample = []
while len(sample)<=100: sample.append(random.randrange(10, 15, 1)*0.5+10)
```

```python
sample
```

```python
plt.plot(sample)
plt.ylim(0, 100)
```

```python
smean_list = [sum(random.choices(sample, k=(i+1)))/(i+1) for i in range(len(sample)+1)]
plt.hist(smean_list, bins=5)
```
