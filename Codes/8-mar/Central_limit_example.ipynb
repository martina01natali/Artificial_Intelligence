# Central limit theorem

- produce N independent random variables distributed with a flat distribution
- compute the sample's mean for k<N variables
- show the "gaussianization" of the mean when k grows

```python
import numpy as np
import random
import matplotlib
from matplotlib import pyplot as plt
```

```python
sample = []
while len(sample)<=100: sample.append(random.randrange(10, 15, 1)*0.5+10)
```

```python
plt.plot(sample)
plt.ylim(0, 100)
```

```python
for length in range(1,len(sample)+1,5):
    smean_list = [sum(random.choices(sample, k=(i+1)))/(i+1) for i in range(length)]
    plt.figure()
    plt.hist(smean_list, bins=int(np.sqrt(len(smean_list))))
```
